## READING INVENTORY.JSON FILE AND MENU FUNCTION (1-time run)

```python
# READING INVENTORY.JSON FILE
import json
import time

fd = open("Inventory.json","r")
r = fd.read()
fd.close()

Inventory = json.loads(r)



# MENU FUNCTION
def menu():
  print("                MENU")
  print("------------------------------------")

  for i in range(1,len(Inventory)+1):
    print()
    print(" PRODUCT ID            : ",int(i))
    print(" PRODUCT NAME          : ",Inventory[str(i)]['NAME'])
    print(" PRODUCT BRAND         : ",Inventory[str(i)]['BRAND'])
    print(" PRODUCT CATEGORY      : ",Inventory[str(i)]['CATEGORY'])
    print(" PRODUCT PRICE         : ",Inventory[str(i)]['SELL_PRICE'])
    print(" PRODUCT QUANTITY LEFT : ",Inventory[str(i)]['QUANTITY_IN_STOCK'])

  print("\n\n")
  print("Do you want to buy anything?")
  print("(a) YES")
  print("(b) NO")

sales = {}
bill_num = 0
```

## PURCHASING ITEMS AND BILLING (Multiple Times Run, 1-time per customer)

```python
# SHOWING MENU FOR THE CUSTOMER AND PURCHASING OF ITEMS
menu()
user_choice = input("\nEnter Your Choice : ")

total_items = 0
item_number = []
item_quantity = []

while (user_choice != 'b' and user_choice != 'B'):
  if (user_choice == 'a' or user_choice == 'A'):
    ui_prod = str(input("\nEnter The Product ID Of Your Item : "))
    if (((int(ui_prod)) > 0) and ((int(ui_prod)) <= len(Inventory))):
      ui_quant = int(input("Enter The Number Of Units You Want To Order For This Item : "))
      if (ui_quant <= int(Inventory[ui_prod]["QUANTITY_IN_STOCK"])):
        total_items += 1
        item_number.append(ui_prod)
        item_quantity.append(ui_quant)
        Inventory[ui_prod]["QUANTITY_IN_STOCK"] = Inventory[ui_prod]["QUANTITY_IN_STOCK"] - ui_quant   
      else:
        print("\nEntered Quantity For The Item Is Not In Stock. Please Try Again.")
    else:
      print("\nEntered Product ID Is Not Valid!")
  else: 
    print("\nINVALID INPUT!")
  
  print("\n\n\nDo you want to buy anything?")
  print("(a) YES")
  print("(b) NO")
  user_choice = input("\nEnter your choice : ")



# BILLING
discount = 0.00
total_amount = 0.00
if (total_items != 0):
  bill_num += 1

  customer_name = str(input("\nEnter The Customer's Name : "))
  customer_phone_number = int(input("Enter The Customer's Phone Number (10 digits) : "))
  
  n = customer_phone_number
  digits_in_phone_number = 0
  while (n != 0):
    n = n//10
    digits_in_phone_number += 1

  if (digits_in_phone_number != 10):
    print("Invalid Phone Number!")
    customer_phone_number = int(input("\nEnter The Customer's Phone Number Again! (10 digits) : "))

    n = customer_phone_number
    digits_in_phone_number = 0
    while (n != 0):
      n = n//10
      digits_in_phone_number += 1
    if (digits_in_phone_number != 10):
      print("Invalid Phone Number Again! Not Added In Bill!")

  print("\n\n\n                       WALMART")
  print("        Store No.- 28, Laxmi Nagar, Delhi - 110092")
  print("=========================================================")
  print("\n Customer Name:     ",customer_name)
  if (digits_in_phone_number == 10):
    print(" Phone Number:      ",customer_phone_number)
  print(" Bill Number:       ",bill_num)
  time_india = time.time() + 19800
  print(" Timestamp Of Bill: ",time.ctime(time_india))
  print()
  print("==========================================================")
  
  for i in range(total_items):
    print()
    print(" PRODUCT NAME             : ",Inventory[item_number[i]]['NAME'])
    print(" PRODUCT BRAND            : ",Inventory[item_number[i]]['BRAND'])
    print(" PRODUCT QUANTITY ORDERED : ",item_quantity[i])
    print(" PRODUCT PRICE PER UNIT   : ",Inventory[item_number[i]]['SELL_PRICE'])
    print(" PRODUCT NET PRICE        : ",item_quantity[i] * (Inventory[item_number[i]]['SELL_PRICE']))
    total_amount += ((Inventory[item_number[i]]['SELL_PRICE']) * item_quantity[i])
    
    if (digits_in_phone_number == 10):
      sales[len(sales) + 1] = {"CUSTOMER_NAME": customer_name, "CUSTOMER_PHONE_NUMBER": customer_phone_number, "PRODUCT_ID": ui_prod, "PRODUCT_NAME": Inventory[item_number[i]]['NAME'], "PRODUCT_BRAND": Inventory[item_number[i]]['BRAND'], "PRODUCT_QUANTITY_ORDERED": item_quantity[i], "PRODUCT_SELL_PRICE_PER_UNIT": Inventory[item_number[i]]['SELL_PRICE'], "PRODUCT_NET_SALE": item_quantity[i] * (Inventory[item_number[i]]['SELL_PRICE']), "TIMESTAMP": time.ctime(time_india), "BILL_NUMBER": bill_num}
    else:
      sales[len(sales) + 1] = {"CUSTOMER_NAME": customer_name, "CUSTOMER_PHONE_NUMBER": 'Not Provided', "PRODUCT_ID": ui_prod, "PRODUCT_NAME": Inventory[item_number[i]]['NAME'], "PRODUCT_BRAND": Inventory[item_number[i]]['BRAND'], "PRODUCT_QUANTITY_ORDERED": item_quantity[i], "PRODUCT_SELL_PRICE_PER_UNIT": Inventory[item_number[i]]['SELL_PRICE'], "PRODUCT_NET_SALE": item_quantity[i] * (Inventory[item_number[i]]['SELL_PRICE']), "TIMESTAMP": time.ctime(time_india), "BILL_NUMBER": bill_num}

  print("==========================================================")
  print()
  print(" Billing Amount                           Rs.",total_amount)
  if (total_amount > 5000.00):
    discount = 0.05 * total_amount
  elif (total_amount > 10000.00):
    discount = 0.1 * total_amount
  else:
    discount = 0.00
  print(" Discount                                 Rs.",discount)
  print(" Net Pay                                  Rs.",total_amount - discount)
  print("==========================================================")



# UPDATING INVENTORY.JSON AFTER PURCHASE
js = json.dumps(Inventory)

fd = open("Inventory.json","w")
fd.write(js)
fd.close()



# STORING SALES DETAILS IN SALES.JSON FILE
sale = json.dumps(sales)

fd = open("sales.json","w")
r = fd.write(sale)
fd.close()
```
